```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('tabby.csv')
```

```python
df['baseline_percent'] = round(df["baseline"]/df["Records"]*100, 2)
df['bm25_percent'] = round(df["bm25"]/df["Records"]*100, 2)
df['oracle_percent'] = round(df['oracle']/df["Records"]*100, 2)
```

```python
df.head()
```

```python
for dataset, sdf in df.groupby('Dataset'):
  fig, ax = plt.subplots()
  ax = sdf.plot(x='Model', y=['baseline_percent', 'bm25_percent', 'oracle_percent'], kind='barh', ax=ax)
  ax.set_title(dataset)
    #ax = pdf.plot(x='Model', y=['% baseline'], kind='barh', ax=ax)
    #df.plot.barh(x='Model', y='% baseline')
plt.plot()
```
